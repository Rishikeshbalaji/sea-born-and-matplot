```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 

```

```python
df= pd.read_csv('cars.csv')
df
```

```python
plt.plot(df['Weight'], df.Origin)
plt.show()
```

```python
df.corr()
```

```python
# heat mam seaborn
sns.heatmap(df.corr())
```

```python
# color palettes.
sns.heatmap(df.corr(), annot=True, cmap='rocket')
```

```python
# removing Cylinders and Year vlaues fom the data set since it not useful to analyse the data set 
data=df.drop(columns=['Cylinders','Year'])
```

```python
data.corr()
```

```python
sns.heatmap(data.corr(), annot=True, cmap='rocket')
plt.figure(figsize=(10,5))
plt.show()
```

```python
# sea born barplot of Engine and Horsepower
sns.barplot(df.Engine_Disp, df.Horsepower, data=df)
plt.show()
```

```python
# draw a bar plot with Accelerate  and origin, year with one as hue
sns.barplot(y=df.Accelerate, x=df.Origin, data=df, palette='Blues',hue='Origin')
plt.show()
```

```python
# draw a bar plot with Accelerate  and Year, year with one as hue
sns.barplot(y=df.Weight, x=df.Year, data=df, palette='Blues', hue='Origin')
plt.show()
```

```python
# draw a countplot for each variable model, origin year 
sns.countplot(x="Origin",hue="Model", data=df)
plt.show()

```

```python
sns.countplot(x='Model',hue='Origin', data=df)
plt.show()
```

```python
sns.countplot(x='Model',hue='Year', data=df)
plt.show()
```

```python
sns.countplot(x='Year',hue='Origin', data=df)
plt.show()
```

```python
sns.boxplot(x=df.Accelerate,y= df.Engine_Disp)
```

```python
# stripplot used to find the dencity 
sns.stripplot(x=df.Year, y=df.Engine_Disp, hue='Origin',data=df)
```

```python
# join plot 
sns.jointplot(df.Horsepower,df.Accelerate ,data=df, space=0, color="b")                                    
plt.show()
```
