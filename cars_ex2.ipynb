```python
import numpy as np
import pandas as pd
from scipy import stats, integrate
import matplotlib.pyplot as plt

import seaborn as sns

%matplotlib inline
```

```python
sns.set(color_codes=True)
```

```python
df=pd.read_csv('cars.csv')
df
```

```python
# box plot
sns.boxplot(df['Actual_MPG'])
plt.show()
```

```python
# violinplot
sns.violinplot(df['Actual_MPG'])
plt.show()
```

```python
# seting figer size
sns.displot(df['Actual_MPG'],height=8, aspect=11/8)
plt.show()
```

```python
sns.displot(df['Actual_MPG'], kde= False, bins=20, rug=True)
plt.show()
```

```python
sns.displot(df['Actual_MPG'], kde= True, bins=20, rug=True)
plt.show()
```

```python
sns.jointplot(df['Actual_MPG'],df['Horsepower'] ,data=df, space=0, color="b")                                    
plt.show()
```

```python
sns.jointplot(df['Actual_MPG'],df['Horsepower'] ,data=df, space=0, color="b", )   
```

```python
sns.jointplot(y= df['Actual_MPG'], x= df['Horsepower'], kind='kde', data= df, cset =8)
plt.show()
```

```python
df.corr()
sns.pairplot(data=df)
plt.show()
```

```python
sns.pairplot(data=df, kind='kde')
plt.show()
```
