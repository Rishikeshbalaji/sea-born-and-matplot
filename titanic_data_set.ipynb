```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.jointplot(titanic.fare,titanic.age ,data=titanic, space=0, color="b")                                    
plt.show()
```

```python
sns.displot(x=titanic.fare)
plt.show()
```

```python
sns.boxplot(x=titanic['class'], y=titanic.age, data=titanic)
plt.show()
```

```python
sns.catplot(x='class', y='age',data=titanic,kind="swarm")
plt.show()
```

```python
sns.displot(titanic.sex, color='b' )
```

```python
titanic.corr()
```

```python
sns.heatmap(titanic.corr(),annot=True)
```

```python
sns.displot( data=titanic,  x='age',  hue='sex',col='sex',fill=True)
```
